```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# faire un graphique qui monte en animation genre l’acceleration des véhicules dnas FZERO

x:np.ndarray = np.arange(-20,21,0.5)
y = np.sqrt(x+20)*3

plt.plot(x, y, color='#004488')
plt.fill(x, y)
plt.axis('scaled')
plt.show()
```
