```python
# histogrammes
import numpy as np
import matplotlib.pyplot as plt
```

```python
val = np.random.randn(1000)

plt.plot(val, 'o')
```

Pour faire un histogramme, on définit un set d’intervalles et on compte le nombre de valeurs dans ces derniers

```python
def n_in_interval(arr, bot, top) -> int:
    count = 0
    for x in arr :
        if x >= bot and x < top : count += 1
    return count


def histo(data, start, finish, cols) -> dict:
    res = {
        'histo': []
    }
    bars = np.linspace(start, finish, cols)
    res['intervals'] = bars
    for i,n in enumerate(bars[:len(bars)-1]) :
        res['histo'].append( n_in_interval(data, bars[i], bars[i+1]) )

    return res

histogramme = histo(val, -4, 4, 20)
print(histogramme['histo'])
print(histogramme['intervals'])
```

```python
def calc_intervals(arr):
    res = list()
    for i,x in enumerate(arr[:len(arr)-1]):
        res.append(round(
            (arr[i] + arr[i+1]) / 2,
            2
            ))

    return res

print(calc_intervals(histogramme['intervals']))

plt.bar(
    calc_intervals(histogramme['intervals']),
    histogramme['histo'],
    # linewidth=0.8,
    edgecolor="white",
    width=0.3
)
```
