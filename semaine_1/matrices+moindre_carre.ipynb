# MOINDRE CARRÉ

```python
import numpy as np

def affine(a,b,x1,x2):
  y = list()
  xx = list()
  x = range(x1,x2)
  for n in x :
    y.append(a*n+b)
    xx.append(n)
  return np.array(y), xx

a_reel = -.15
b =  0
x1 = -10
x2 = 10
y,x = affine(a_reel,b,x1,x2)

import matplotlib.pyplot as plt

plt.plot(x,y,'-o')
plt.grid()
```

```python
# génération d’une erreur de mesure selon une loi normale
yerr = y + np.random.randn(y.shape[0])/5
```

```python
plt.plot(x,y)
plt.plot(x,yerr,'o')
plt.grid()
```

```python
def erreur_model(verite,prediction):
  err = np.mean( (verite-prediction)**2  )
  return err

def erreur_observation(observation,prediction):
  err = np.mean( (observation-prediction)**2  )
  return err

list_err = list()
list_obs = list()
list_a = np.arange(-2,2,.01)
for a in list(list_a):
  y_test,x = affine(a,0,x1,x2)
  list_err.append(erreur_model(y,y_test))
  list_obs.append(erreur_model(yerr,y_test))


a_est = list_a[np.argmin(list_obs)]
a_reel_2 = list_a[np.argmin(list_err)]

plt.plot(list_a,list_err, label="model/prediction")
plt.plot(list_a,list_obs, label="observation/prediction")
plt.plot(a_est, list_obs[np.argmin(list_obs)],'or')
plt.plot(a_reel_2, list_err[np.argmin(list_err)],'db')
plt.title("Estimation de a {} / {}".format(a_est, a_reel))
plt.grid()
plt.legend()
```

```python
y1,_ = affine(a_reel,0,x1,x2)
y2,_ = affine(a_est,0,x1,x2)
plt.plot(y1, label="true")
plt.plot(y2, label='est')
plt.grid()
plt.legend()
```
