```python
# histogrammes
import numpy as np
import matplotlib.pyplot as plt
```

```python
val = np.random.randn(100)

plt.plot(val, 'o')
```

Pour faire un histogramme, on définit un set d’intervalles et on compte le nombre de valeurs dans ces derniers

```python
def n_in_interval(arr, bot, top) -> int:
    count = 0
    for x in arr :
        if x >= bot and x < top : count += 1
    return count


def hist(data, start, finish, cols) -> list:
    hist = list()
    bars = np.linspace(start, finish, cols)
    for i,n in enumerate(bars[:len(bars)-1]) :
        hist.append( n_in_interval(data, bars[i], bars[i+1]) )

    return hist

histogramme = hist(val, -4, 4, 10)
histogramme
```

```python
# Tracer une courbe sinusoïdale en utilisant NumPy.

x = np.arange(-20,21,0.2)

plt.plot(x,np.sin(x))
plt.ylim(-2.5,2.5)
```

```python
# Tracer deux courbes sur le même graphe à partir des données suivantes :
# x = np.linspace(0, 10, 100)
# y1 = np.sin(x)
# y2 = np.cos(x)

x = np.linspace(0, 10, 50)

plt.plot(x, np.sin(x), '--')
plt.plot(x, np.cos(x), color="orange")
plt.ylim(-2,2)
plt.grid()
```

```python
# Tracer un diagramme en secteurs (camembert) à partir des données suivantes :
# categories = ['A', 'B', 'C', 'D'], valeurs = [30, 25, 15, 20]

categories = ['A', 'B', 'C', 'D']
valeurs = [30, 25, 15, 20]

plt.pie(
    valeurs, 
    labels=categories, 
    startangle=90,
    colors=["gray", "pink", "beige", "peachpuff"],
    explode=[0,0,0.1,0]
    );

# le point virgule permet de ne pas avoir un retour sur la liste des valeurs du diagramme
```

```python
# Tracer une courbe polynomiale à partir des données suivantes :
# x = np.linspace(-5, 5, 100)
# y = 3 * x**3 + 2 * x**2 - 8 * x + 5

x = np.linspace(-5, 5, 100)
y = 3 * x**3 + 2 * x**2 - 8 * x + 5

plt.plot(x,y)
```

```python
# Tracer un diagramme en barres empilées à partir des données suivantes :
# categories = ['A', 'B', 'C', 'D']
# valeurs1 = [10, 15, 20, 25], valeurs2 = [5, 10, 15, 20]

categories = ['A', 'B', 'C', 'D']
valeurs1 = [10, 15, 20, 25]
valeurs2 = [5, 10, 15, 20]

plt.barh(categories, valeurs1)
plt.barh(categories, valeurs2);
```
